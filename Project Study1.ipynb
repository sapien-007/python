### Exploratory Data Analysis

```python
corr = df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig,ax =  plt.subplots(figsize=(15,9))
sns.heatmap(corr, annot=False, ax=ax, cmap='coolwarm', linewidths=0.1)
```

```python
import numpy as np
corr_bool= np.greater_equal(abs(corr),0.75)
corr[np.less(corr[corr_bool],1)]
```

```python
columns_day = df.columns.values[31:38]
columns_day
```

```python
unpop = df[df['shares']<1400]
pop   = df[df['shares']>1400]
```

```python
unpop_day = unpop[columns_day].sum().values
pop_day = pop[columns_day].sum().values
unpop_day
```

```python
pop_visual = pd.DataFrame(data=pop_day, columns=['popsum'],index=columns_day)
unpop_visual = pd.DataFrame(data=unpop_day, columns=['unpopsum'],index=columns_day)
allpop_visual = pop_visual.join(unpop_visual)
allpop_visual
```

```python
allpop_visual.plot.bar()
plt.show()
```

```python
columns_style = df.columns.values[13:19]
columns_style
```

```python
unpop_chan = unpop[columns_style].sum().values
pop_chan = pop[columns_style].sum().values
```
